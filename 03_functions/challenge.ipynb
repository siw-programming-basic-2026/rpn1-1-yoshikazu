```python
#FizzBuzz
#3の倍数はFizz 5の倍数はBuzz　両方の倍数はFizzBuzz

for i in range(25):
    if i == 0:
        print(i)
    elif i%3 == 0 and i%5 == 0:
        print("fizzbuzz")
    elif i%3 == 0:
        print("fizz")
    elif i%5 == 0:
        print("buzz")
    else:
        print(i)
```
